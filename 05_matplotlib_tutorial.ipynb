# Matplotlib 기초

## 1. Matplotlib 소개

### Matplotlib이란?
Matplotlib은 Python에서 가장 널리 사용되는 데이터 시각화 라이브러리입니다. 2003년 John Hunter가 개발하기 시작했으며, MATLAB의 plotting 기능을 Python에서 구현하는 것이 목표였습니다.

### 주요 특징
- **다양한 그래프 유형**: 선 그래프, 막대 그래프, 히스토그램, 산점도 등
- **높은 커스터마이징**: 색상, 스타일, 폰트 등 모든 요소 조정 가능
- **출판 품질**: 논문, 보고서에 사용할 수 있는 고품질 그래프 생성
- **다양한 백엔드**: GUI 창, 웹, 파일 저장 등 다양한 출력 방식 지원
- **NumPy 연동**: NumPy 배열과 완벽 호환

### 활용 분야
- 과학적 연구 및 논문
- 비즈니스 데이터 분석
- 머신러닝 결과 시각화
- 웹 애플리케이션 차트
- 교육 및 프레젠테이션

### 선 그래프
선 그래프: x 축 대비 y축 데이터의 추이, 누적, 변화 확인  
```python
plt.plot(x, y)
```

````python
# 간단한 선 그래프
# 점 (1, 1), (2, 3), (3, 8) 을 선으로 연결하는 그래프
# 따라서, x축과 y 축의 데이터 개수는 동일해야 함.
x = [1, 2, 3]   # x 축 데이터를 기입 (1, 2, 3), x 축 데이터를 기입하지 않으면 자동으로 0부터 1씩 증가하는 데이터가 생성됨.
y = [1, 3, 8]   # y 축 데이터를 기입 (1, 3, 8)
# y = [1, 3, 8, 9] # 에러 발생생

# plt.figure 을 따로 하지 않아도 자동으로 설정됨.
plt.plot(x, y)
plt.title('선 그래프')
plt.xlabel('X축')
plt.ylabel('Y축')
plt.show()
````

### Numpy 배열 그래프

````python
import numpy as np
# NumPy 배열을 사용한 더 복잡한 그래프
x = np.linspace(0, 2*np.pi)  # 0부터 2π까지 100개 점
y = np.sin(x)

plt.figure(figsize=(10, 6))  # 그래프 크기 설정
plt.plot(x, y, linewidth=2, color='blue')
plt.title('사인 함수 그래프', fontsize=16)
plt.xlabel('x (라디안)', fontsize=12)
plt.ylabel('sin(x)', fontsize=12)
plt.grid()
plt.show()


##### pyplot 에서 자주 사용하는 옵션
# grid: 격자 표시
# title: 제목
# xlabel: x축 레이블
# ylabel: y축 레이블
# legend: 범례 표시, loc='upper right', 'upper left', 'lower right', 'lower left' 위치 지정 가능
# show: 그래프 표시
````

### 여러 데이터 시리즈

````python
# 한 그래프에 여러 선 그리기
# 서로 다른 데이터는 자동으로 선 색깔이 달라짐.
# 선끼리 구분하기 위해서 label 옵션을 사용함.
x = np.linspace(0, 2*np.pi)
y1 = np.sin(x)
y2 = np.cos(x)
plt.plot(x, y1, label='sin(x)')
plt.plot(x, y2, label='cos(x)')
plt.legend()
plt.show()
````

## 4. 그래프 스타일링

### 색상과 마커

https://matplotlib.org/stable/gallery/

**plot 에서 자주 사용하는 옵션**  
- linewidth: 선의 두께  
- linestyle: 선의 스타일 ('.', '-', '--', '-.', ':')  

<p align="center"> <img src=https://matplotlib.org/stable/_images/sphx_glr_linestyles_001_2_00x.png width=500 align='center'> </p>

- color: 선의 색상 

<p align="center"> <img src="https://matplotlib.org/stable/_images/sphx_glr_named_colors_001_2_00x.png" width=500 align='center'> </p>
<p align="center"> <img src="https://matplotlib.org/stable/_images/sphx_glr_named_colors_003_2_00x.png" width=500 align='center'> </p>

- marker: 점의 모양
<p align="center"> <img src="https://matplotlib.org/stable/_images/sphx_glr_marker_reference_001_2_00x.png" width=500 align='center'> </p>

- markersize: 점의 크기  

- alpha: 투명도 (0.0 ~ 1.0)  

    

````python
# 다양한 색상과 마커 스타일
x = np.arange(10)
y1 = x
y2 = x * 2
y3 = x * 3

# 그래프 그리기
plt.figure(figsize=(12, 8))
plt.plot(x, y1, color='r', marker='o', markersize=5, label='x')
plt.plot(x, y2, color='g', marker='s', markersize=8, label='2x')
plt.plot(x, y3, color='b', marker='d', markersize=10, label='3x')

plt.title('다양한 색상과 마커', fontsize=16)     # 글씨 크기 가능
plt.xlabel('X값')
plt.ylabel('Y값')
plt.legend() # loc='upper right', 'upper left', 'lower right', 'lower left' 위치 지정 가능
plt.grid(alpha=0.3, linestyle='--', axis='x') # 선처럼 옵션 선택 가능, axis='x', 'y', 'both' 축 지정 가능
plt.show()
````

### 선 스타일과 투명도

````python
# 선 스타일 옵션
x = np.linspace(0, 2*np.pi)

plt.figure(figsize=(12, 10))

# 다양한 선 스타일
plt.plot(x, np.sin(x), linestyle='-', linewidth=3, label='실선, solid (-)', alpha=0.8)
plt.plot(x, np.cos(x), linestyle='--', linewidth=3, label='점선, dotted (--)', alpha=0.8)
plt.plot(x, np.sin(x+np.pi), linestyle='-.', linewidth=3, label='점-대시, dashdot (-.)', alpha=0.8)
plt.plot(x, np.cos(x+np.pi), linestyle=':', linewidth=3, label='점점선, dashed (:)', alpha=0.8)

plt.title('선 스타일 비교', fontsize=16)
plt.xlabel('X값')
plt.ylabel('sin(x) and cos(x)')
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()
````

## 5. 다양한 그래프 유형

### 막대 그래프 (bar graph)

막대 그래프: 범주별 데이터의 분포, 크기를 확인  
```python
plt.bar(x, height)
```

````python
import pandas as pd
df = pd.DataFrame({'팀': ['A', 'B', 'C', 'D'], '점수': [23, 45, 56, 78]})

plt.figure(figsize=(10, 6))
bars = plt.bar(df['팀'], df['점수'])

plt.title('막대 그래프', fontsize=16)
plt.xlabel('팀')
plt.ylabel('점수')
plt.grid()
plt.show()
````

### 히스토그램(Histogram)

히스토그램: 수치형 데이터의 분포 형태, 치우침, 이상치를 확인
```python
plt.hist(data, bins=10)
```

#### 수치형 데이터의 분포 확인
- 중심(Center): 데이터는 어느 값을 중심으로 몰려있는가, 최빈값 또는 분포의 중심을 찾기  
- 범위(Spread): 데이터가 얼마나 퍼져있는가, 분산이 크면 데이터가 넓게 분포되어 있음  
- 모양(Shape): 종모양, 여러 개의 봉우리, 정규분포나 이봉분포 등 확인  
- 치우침(Skewness): 데이터가 어떻게 분포되어 있는가, 좌우 대칭인지, 오른쪽으로 치우친 형태인지 등  
- 이상치(Outliers): 극단적인 값이 존재하는가, 멀리 떨어진 구간에 홀로 있는 막대 탐색  

````python
np.random.seed(42)
data = np.random.normal(100, 15, 1000)  # 평균 100, 표준편차 15인 정규분포

plt.figure(figsize=(10, 6))
plt.hist(data, bins=30, color = 'skyblue', edgecolor = 'black')     # bins: 데이터를 몇 개의 구간으로 나눌 지 조정
# plt.axvline(np.mean(data), color='red', linestyle='--', linewidth=2, label=f'평균: {np.mean(data):.1f}')
# plt.axvline(np.median(data), color='green', linestyle='--', linewidth=2, label=f'중앙값: {np.median(data):.1f}')

plt.title('히스토그램', fontsize=16)
plt.xlabel('점수')
plt.ylabel('빈도')
plt.grid(True, alpha=0.3)
plt.show()
````

### 산점도 (Scatter Plot)

두 변수의 관계, 상관성, 이상치 확인

```python
plt.scatter(x, y)
```

````python
# 산점도
np.random.seed(42)
n = 100
x = np.arange(100)
y = 2*x + np.random.normal(0, 10, n)
colors = y
sizes = 40

plt.figure(figsize=(10, 8))
plt.scatter(x, y, c=colors, s=sizes, alpha=0.6)

# # 추세선 추가
# z = np.polyfit(x, y, 1)
# p = np.poly1d(z)
# plt.plot(x, p(x), "r--", alpha=0.8, linewidth=2, label=f'추세선: y={z[0]:.2f}x+{z[1]:.2f}')

plt.title('산점도', fontsize=16)
plt.xlabel('X값')
plt.ylabel('Y값')
# plt.legend()
plt.grid(True, alpha=0.3)
plt.show()
````

### 박스 플롯

수치형 데이터의 중앙값/사분위/이상치 비교

```python
plt.boxplot(data)
```

#### 분포의 요약 통계

- 박스 중앙선: 중앙값 (50번째 백분위수)
- 박스 아래쪽: 25번째 백분위수
- 박스 위쪽: 75번째 백분위수
- 수염(whisker): 1.5×IQR 범위
- 점들: 이상치(outliers)

````python
np.random.seed(42)
data = np.random.normal(100, 10, 100)
plt.figure(figsize=(10, 8))
box_plot = plt.boxplot(data)

plt.title('박스 플롯', fontsize=16)
plt.xlabel('그룹')
plt.ylabel('점수')
plt.grid(True, alpha=0.3)
plt.show()
````

````python
np.random.seed(42)
data = [np.random.normal(100, 10, 100),
        np.random.normal(110, 15, 100),
        np.random.normal(95, 12, 100),
        np.random.normal(105, 8, 100)]

labels = ['그룹A', '그룹B', '그룹C', '그룹D']

plt.figure(figsize=(10, 8))
plt.boxplot(data, tick_labels=labels)

plt.title('박스 플롯', fontsize=16)
plt.xlabel('그룹')
plt.ylabel('점수')
plt.grid(True, alpha=0.3)
plt.show()
````

### 그래프 유형별 사용 가이드

| 그래프 유형 | 언제 사용? | 데이터 유형 | 예시 |
|------------|-----------|------------|------|
| 선 그래프 | 시간에 따른 변화 추이 | 연속형 (시계열) | 주가 변화, 기온 변화 |
| 막대 그래프 | 범주별 크기 비교 | 범주형 + 연속형 | 지역별 매출, 제품별 판매량 |
| 히스토그램 | 데이터 분포 확인 | 연속형 | 시험 점수 분포, 키 분포 |
| 산점도 | 두 변수 간 관계 | 연속형 + 연속형 | 키와 몸무게, 공부시간과 성적 |
| 파이 차트 | 전체 대비 비율 | 범주형 | 시장 점유율, 예산 구성 |
| 박스 플롯 | 분포 요약 통계 | 연속형 | 그룹별 성능 비교 |

## 6. 서브플롯과 레이아웃

### 기본 서브플롯

````python
x = np.linspace(0, 2*np.pi, 100)

# 2x2 서브플롯

# fig, axes = plt.subplots(2, 2, figsize=(5, 4))
# ax1, ax2, ax3, ax4 = axes.ravel()

fig = plt.figure(figsize=(5, 4))
ax1 = fig.add_subplot(221)
ax2 = fig.add_subplot(222)
ax3 = fig.add_subplot(223)
ax4 = fig.add_subplot(224)


ax1.plot(x, np.sin(x))
ax1.set_title('sin(x)')
ax1.grid(True, alpha=0.3)

ax2.plot(x, np.cos(x), 'r-')
ax2.set_title('cos(x)')
ax2.grid(True, alpha=0.3)

ax3.plot(x, np.tan(x))
ax3.set_title('tan(x)')
ax3.set_ylim(-5, 5)
ax3.grid(True, alpha=0.3)

ax4.plot(x, np.sinh(x))
ax4.set_title('sinh(x)')
ax4.grid(True, alpha=0.3)

plt.tight_layout()  # 서브플롯 간격 자동 조정
plt.show()
````

## 7. 고급 시각화 기법

### 3D 그래프

````python
n = 100
x = np.random.randn(n)
y = np.random.randn(n)
z = np.random.randn(n)
colors = np.random.rand(n)

# 3D 산점도
fig = plt.figure(figsize=(6, 5))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(x, y, z, c=colors, marker='o')
ax.set_title('3D Scatter Plot')
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
plt.show()
````
